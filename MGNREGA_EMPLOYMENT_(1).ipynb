## 1. Correlation Analysis:

We started by performing a correlation analysis between the attributes 'EMPLOYMENT DEMANDED,' 'EMPLOYMENT OFFERED,' and 'EMPLOYMENT AVAILED.'

```python
import pandas as pd
import numpy as np

# Read the data from the Excel file into a DataFrame
  # Replace 'path_to_your_excel_file.xlsx' with the actual file path
df = pd.read_excel('employment_analysis.xlsx',sheet_name='national')

# Calculate the Pearson correlation coefficients
correlation_matrix = df.corr(method='pearson')

# The correlation between 'EMPLOYMENT DEMANDED' and 'EMPLOYMENT OFFERED'
correlation_demanded_offered = correlation_matrix.loc[:,['employment demanded', 'employment offered']]

# The correlation between 'EMPLOYMENT DEMANDED' and 'EMPLOYMENT AVAILED'
correlation_demanded_availed = correlation_matrix.loc[:,['employment demanded', 'employment availed']]

# The correlation between 'EMPLOYMENT OFFERED' and 'EMPLOYMENT AVAILED'
correlation_offered_availed = correlation_matrix.loc[:,['employment offered', 'employment availed']]

print("Correlation between 'EMPLOYMENT DEMANDED' and 'EMPLOYMENT OFFERED':\n", correlation_demanded_offered)
print("Correlation between 'EMPLOYMENT DEMANDED' and 'EMPLOYMENT AVAILED':\n", correlation_demanded_availed)
print("Correlation between 'EMPLOYMENT OFFERED' and 'EMPLOYMENT AVAILED':\n", correlation_offered_availed)
```

The Pearson correlation coefficients for the pairs of attributes are as follows:

Correlation between 'EMPLOYMENT DEMANDED' and 'EMPLOYMENT OFFERED': 0.999959
Correlation between 'EMPLOYMENT DEMANDED' and 'EMPLOYMENT AVAILED': 0.999449
Correlation between 'EMPLOYMENT OFFERED' and 'EMPLOYMENT AVAILED': 0.999236
The high positive correlations between 'EMPLOYMENT DEMANDED' and both 'EMPLOYMENT OFFERED' and 'EMPLOYMENT AVAILED' indicate a strong linear relationship. It suggests that there is a close association between the number of employment opportunities demanded by rural households and the number of opportunities offered and availed under the MGNREGA scheme. This finding suggests that the program has been successful in fulfilling the employment demands of the rural population.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data from Excel file using pandas
data = pd.read_excel("employment_analysis.xlsx", engine="openpyxl")

# Calculate the correlation matrix
corr_matrix = data.corr()

# Define color scheme for heatmap (you can change the cmap parameter to different color palettes)
cmap = sns.diverging_palette(220, 20, as_cmap=True)

# Plot the heatmap using seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap=cmap, vmin=-1, vmax=1)
plt.title("Correlation Matrix Heatmap")
plt.show()

```

## 2. Scatter Plot Analysis:

We created scatter plots to visualize the relationships between the pairs of attributes, 'EMPLOYMENT DEMANDED' and 'EMPLOYMENT OFFERED,' and 'EMPLOYMENT DEMANDED' and 'EMPLOYMENT AVAILED,' and 'EMPLOYMENT OFFERED' and 'EMPLOYMENT AVAILED.'
The scatter plots provide a visual representation of the correlation between employment demanded, offered, and availed, amplifying the results of the correlation analysis.

```python

import matplotlib.pyplot as plt
%matplotlib inline
# Read the data from the Excel file into a DataFrame
file_path = 'employment_analysis.xlsx'  # Replace 'path_to_your_excel_file.xlsx' with the actual file path
df = pd.read_excel(file_path)
employment_demanded = df['employment demanded'].values
employment_offered = df['employment offered'].values
employment_availed = df['employment availed'].values
```

```python
# Scatter plot: EMPLOYMENT DEMANDED vs. EMPLOYMENT OFFERED
slope, intercept = np.polyfit(employment_demanded, employment_offered, 1)
plt.scatter(df['employment demanded'], df['employment offered'])
plt.plot(employment_demanded, slope * employment_demanded + intercept, color='red')
plt.xlabel('EMPLOYMENT DEMANDED')
plt.ylabel('EMPLOYMENT OFFERED')
plt.title('Scatter Plot: EMPLOYMENT DEMANDED vs. EMPLOYMENT OFFERED')
plt.show()
```

```python
# Scatter plot: EMPLOYMENT DEMANDED vs. EMPLOYMENT AVAILED
slope, intercept = np.polyfit(employment_demanded, employment_availed, 1)
plt.scatter(df['employment demanded'], df['employment availed'])
plt.plot(employment_demanded, slope * employment_demanded + intercept, color='red')
plt.xlabel('EMPLOYMENT DEMANDED')
plt.ylabel('EMPLOYMENT AVAILED')
plt.title('Scatter Plot: EMPLOYMENT DEMANDED vs. EMPLOYMENT AVAILED')
plt.show()
```

```python
# Scatter plot: EMPLOYMENT OFFERED vs. EMPLOYMENT AVAILED
slope, intercept = np.polyfit(employment_offered, employment_availed, 1)
plt.scatter(df['employment offered'], df['employment availed'])
plt.plot(employment_offered, slope * employment_offered + intercept, color='red')
plt.xlabel('EMPLOYMENT OFFERED')
plt.ylabel('EMPLOYMENT AVAILED')
plt.title('Scatter Plot: EMPLOYMENT OFFERED vs. EMPLOYMENT AVAILED')
plt.show()
```

## 3.Trend over time

Next, we explored the trends over time for each of the attributes, 'EMPLOYMENT DEMANDED,' 'EMPLOYMENT OFFERED,' and 'EMPLOYMENT AVAILED.'

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the Excel file into a DataFrame
file_path = 'employment.xlsx'  # Replace 'path_to_your_excel_file.xlsx' with the actual file path
df = pd.read_excel(file_path)

# Create a figure and axis for the plot
fig, ax = plt.subplots()

# Plot the trends for each attribute
ax.plot(df['year'], df['employment demanded'], label='Employment Demanded', marker='o')
ax.plot(df['year'], df['employment offered'], label='Employment Offered', marker='o')
ax.plot(df['year'], df['employment availed'], label='Employment Availed', marker='o')

# Set labels and title
ax.set_xlabel('Year')
ax.set_ylabel('Number of Employments')
ax.set_title('Trends of Employment Demanded, Offered, and Availed over Time')
# Show legend
ax.legend()
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

The line chart demonstrates the changing patterns of employment demanded, offered, and availed through MGNREGA over the years.The employment availed follows a similar trend, closely tracking the employment offered by the scheme. This alignment between the opportunities offered and availed suggests efficient implementation and responsiveness of the MGNREGA program to the beneficiaries' needs.
Here we can see a dip in all the three line plots after the year 2020-2021.One of the reasons speculated is the government cutting the spending to Rs 60,000 crore for the next financial year from Rs 89,400 crore in 2022-23 and also the dues needing to be paid as a result.

## 4. Growth rate(GR) of budget allocated

Here, 2014-2015 data has been considered as the base year.
The formula for gc= (value of present year-value of past year)/value of past year*100

Here we can the fund allocated during the year 2020-2021 has a major growth from the previous year.Whereas after the year 2020-2021 we can see a decline in the budget allocated ,where fund allocated to MGNREGA fell by 32.89% in the year 2023-2024 as compared to 2022-2023.

## 5. Payment due for the prior year

Here we analyse the percentage of MGNREGA fund allocated that has been used for paying the bill due in the latter year.
Formula used for % of fund to pay the due=(fund released-payment due)/fund released*100

After the covid year 2020-2021 we can see a significant increase(5.88%) in the percentage of fund being used to pay the due bill,along with the freezing of budget by the government. Only an increase in the percentage in the subsequent years can be seen.

## 6. Household completed 100 days of employment

Formula used for percentage of household completed 100 days of employment=household completed 100 days/registered household*100

According to the analysis provided above the registered households receiving 100 days of employment is only 3.12% in average, between 2014-2015 and 2022-2023
